```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langgraph.graph import StateGraph,START,END,MessagesState
from dotenv import load_dotenv
from typing import TypedDict
from pprint import pprint
```

```python
load_dotenv()

class BlogState(TypedDict):

    title:str
    outline:str
    content:str
```

```python
model=ChatGoogleGenerativeAI(  model="gemini-3.5-flash",temperature=0)
```

```python
def create_outline(state:BlogState)->BlogState:

    # fetch title
    title=state['title']

    # add prompt and call model
    prompt=f'generate the outline for blog on topic f{title}'
    outline=model.invoke(prompt).content

    #update state

    state['outline']=outline

    return state
```

```python
def create_content(state:BlogState)->BlogState:
    
    # fetch outline and title
    outline=state['outline']
    title=state['title']

    #prompt and model calling
    prompt=f'write a sytematic blog using this outline f{outline} and title {title}'
    content=model.invoke(prompt).content

    #update state

    state['content']=content

    return state
```

```python
graph=StateGraph(BlogState)

#nodes

graph.add_node('outline',create_outline)
graph.add_node('content',create_content)

#edges

graph.add_edge(START,'outline')
graph.add_edge('outline','content')
graph.add_edge('content',END)

#compile

workflow=graph.compile()
```

```python
intial_state={'title':'Rise of ai in india'}

final_state=workflow.invoke(intial_state)

print(final_state)
```
